```python
import pandas as pd

# Specify the path to your CSV file
file_path = f"tweets-data/{filename}"

# Read the CSV file into a pandas DataFrame
df = pd.read_csv(file_path, delimiter=";")

# Display the DataFrame
display(df)
```

Pre-Processing 1

```python
#drop kolom yang tidak dibutuhkan (index kolom dimulai dari nol)
df.drop(df.columns[[0,1,3,4,5,6,7,8,9,11]], axis=1, inplace=True)

display(df)
```

Pre-Processing 2 : Membuat kolom full_text menjadi Lower_Case

```python
def casefoldingText(text): # Converting al the characters in a text into lower case
    text = text.lower()
    return text

df['Lower_Case'] = df['full_text'].apply(casefoldingText)
df.head(10)
```

Pre-Processing 3 : Membersihkan text yang tidak diperlukan dengan menggunakan Regular Expression (Regex)

```python
# Cleaning
import re
import string

def cleaningText(text):
    text = re.sub(r'@[A-Za-z0-9]+', '', text) # remove mentions
    text = re.sub(r'#[A-Za-z0-9]+', '', text) # remove hashtag
    text = re.sub(r'RT[\s]','', text) # remove RT
    text = re.sub(r"http\S+",'', text) # remove link
    text = re.sub(r'[0-9]+','', text) # remove numbers
    text = re.sub(r'[^A-Za-z ]+','', text) # remove all character non alfabet

    text = text.replace('\n',' ') # replace new line into space
    text = text.translate(str.maketrans('', '', string.punctuation)) # remove all punctuations
    text = text.strip(' ') # remove characters space from both left and right text
    return text

df['cleaning'] = df['Lower_Case'].apply(cleaningText)
df.head(20)
```

Pre-Processing 4 : Tokenisasi data untuk memisahkan kalimat menjadi kata per kata

```python
import nltk
nltk.download('punkt')
```

```python
# Tokenisasi
from nltk.tokenize import word_tokenize

def tokenizingText(text): # Toeknizing or splitting a string, text into a list of tokens
    text = word_tokenize(text)
    return text

df['tokenizing'] = df['cleaning'].apply(tokenizingText)
# tweets = tweets[['text_clean', 'text_preprocessed']]
df.head(30)
```
